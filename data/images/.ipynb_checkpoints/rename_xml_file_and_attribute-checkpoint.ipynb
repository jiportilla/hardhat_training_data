```python
# cahn

import os
from shutil import copyfile
import xml.etree.ElementTree as ET
```

```python
# os.chdir('data/images')
attribute='filename'
```

```python
os.mkdir('test_')
os.mkdir('train_')
```

```python
# copy over files
for t in ['test']:
    count=0
    xml=sorted([i for i in os.listdir(t) if '.xml' in i])
    jpg=sorted([i for i in os.listdir(t) if '.jpeg' in i])
    
    for x,j in zip(xml,jpg):
        copyfile(t+'/'+x, t+'_/'+'{}_{}_'.format(t,count)+'.xml')
        copyfile(t+'/'+j, t+'_/'+'{}_{}_'.format(t,count)+'.jpeg')
        
        count+=1
```

```python
# change attribute
for x in xml:
    tree=ET.parse(t+'_/'+x)
    root=tree.getroot()
    for elem in root.iter(attribute):
        elem.text='{}'.format('{}'.format(x.split('.')[0])+'.jpeg')
        #print(elem.text)
        tree.write('{}_/{}'.format(t,x))
```

```python
for t in ['train']:
    count=0
    xml=sorted([i for i in os.listdir(t) if '.xml' in i])
    jpg=sorted([i for i in os.listdir(t) if '.jpeg' in i])
    
    for x,j in zip(xml,jpg):
        copyfile(t+'/'+x, t+'_/'+'{}_{}_'.format(t,count)+'.xml')
        copyfile(t+'/'+j, t+'_/'+'{}_{}_'.format(t,count)+'.jpeg')
        
        count+=1
```
