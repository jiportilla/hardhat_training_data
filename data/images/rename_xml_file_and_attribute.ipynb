```python
# changes file name and attribute

import os
from shutil import copyfile
import xml.etree.ElementTree as ET

attribute='filename'
for t in ['test','train']:
    count=0
    xml=sorted([i for i in os.listdir(t) if '.xml' in i])
    jpg=sorted([i for i in os.listdir(t) if '.jpeg' in i])
    for x,j in zip(xml,jpg):
        copyfile(t+'/'+x, t+'_/'+'{}_{}_'.format(t,count)+'.xml')
        copyfile(t+'/'+j, t+'_/'+'{}_{}_'.format(t,count)+'.jpeg')

        tree = ET.parse(t+'_/'+x)
        root = tree.getroot()
        
        # changing a field text
        for elem in root.iter(attribute):
            elem.text='{}'.format('{}_{}_'.format(t,count)+'.jpeg')
            print(elem.text)
            tree.write('{}_/{}_{}_'.format(t,t,count)+'.xml')
        
        count+=1
```
